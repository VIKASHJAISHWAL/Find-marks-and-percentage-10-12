```python
class Student:

    def __init__(self):
        self.name = ""
        self.clas = ""
        self.subject = {}
        self.total = 0
        self.percentage = 0
        self.grade = ""
        self.menu()

    def menu(self):
        while True:
            choice = input("""
   ============= Student Menu =============
   1. Name and Class
   2. Calculate Total and Percentage
   3. Get Grade
   4. Show Result
   5. Exit
   ========================================
   Enter your choice: """)

            if choice == '1':
                self.Name_and_class()

            elif choice == '2':
                self.Calculate_Percentage()

            elif choice == '3':
                self.Get_Grade()

            elif choice == '4':
                self.Show_Result()

            elif choice == '5':
                print("Thank you! Exiting...")
                break

            else:
                print("Invalid Choice! Try Again.")

    def Name_and_class(self):
        user = input("Enter Student Name: ")
        self.name = user
        clas = input("Enter Class (10/12): ")
        self.clas = clas
        print("Name & Class Saved Successfully!")

    def Calculate_Percentage(self):
        self.subject.clear()

        if self.clas == '10':
            total_subjects = 6
        elif self.clas == '12':
            total_subjects = 5
        else:
            print("Please enter valid class (10 or 12)")
            return

        for i in range(1, total_subjects + 1):
            sub = input(f"Enter {i}th subject name: ")
            mar = int(input(f"Enter marks for {sub}: "))
            self.subject[sub] = mar

        self.total = sum(self.subject.values())
        self.percentage = self.total / total_subjects

        print(f"Total Marks = {self.total}")
        print(f"Percentage = {self.percentage:.2f} %")

    def Get_Grade(self):
        if self.percentage == 0:
            print("Please calculate percentage first!")
            return

        if self.percentage >= 90:
            self.grade = "A+"
        elif self.percentage >= 80:
            self.grade = "A"
        elif self.percentage >= 70:
            self.grade = "B"
        elif self.percentage >= 60:
            self.grade = "C"
        elif self.percentage >= 50:
            self.grade = "D"
        else:
            self.grade = "Fail"

        print(f"Grade = {self.grade}")

    def Show_Result(self):
        print("\n===== STUDENT RESULT =====")
        print(f"Name      : {self.name}")
        print(f"Class     : {self.clas}")
        print("----------- Marks -----------")

        for sub, mark in self.subject.items():
            print(f"{sub} : {mark}")

        print("-----------------------------")
        print(f"Total Marks : {self.total}")
        print(f"Percentage  : {self.percentage:.2f}%")
        print(f"Grade       : {self.grade}")
        print("=============================\n")


obj = Student()
```
